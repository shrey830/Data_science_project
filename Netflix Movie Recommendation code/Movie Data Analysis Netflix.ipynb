```python
# Exploration Summary 

# . We have DataFrame consisting of 9827 rows and 9 columns
# . Our DataSet look a bit tidy with no NaNs or duplicated Values
# . Release Date Column need Casted into date time and to extract only the year value
# . Overview Original_Language and Psoter-Url Wouldn't be So Useful during analysis So We Will drop them
# . there is noticible outliers in Populiarity Column
# . Vote_Average better categorized for analysis
# . Genre column has comma Separated Values and White Spaces that needs to habdle and casted into Category.Exploration Summary
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'])

df['Release_Date'] = df['Release_Date'].dt.year

df.head()
```

```python
# Dropping The Column 

column = ['Overview', 'Original_Language', 'Poster_Url']
df.drop(column, axis = 1, inplace = True)

df.head()
```

```python
# Categorizing Vote_Average Column 

def Categorize_col(df, col, labels):
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df
```

```python
labels = ['Not Popular', 'Below Avg', 'Average', 'Popular']

Categorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```

```python
df.dropna(inplace = True)

df.isna().sum()
```

```python
df['Genre'] = df['Genre'].str.split(', ')

df = df.explode('Genre').reset_index(drop=True)

df.head()
```

```python
df['Genre'] = df['Genre'].astype('category')

df['Genre'].dtypes
```
